```python
import torch 
import cv2
import numpy as np
# x = torch.rand(24,1,18,20)
import torchvision
shape = (96, 96)
image = torchvision.io.read_image('dog.jpg')
image = image.unsqueeze(0)
image = [image for i in range(24)]
image = torch.cat(image, dim=0)
print('image shape = ', image.shape)
from torch.nn import functional as F 
y = F.interpolate(image, shape)
print('y shape = ', y.shape)

# plot a batch of images from torchvision
import matplotlib.pyplot as plt
def plot_batch(batch):
    grid = torchvision.utils.make_grid(batch)
    # plt.figure(figsize=(10,10))
    # plt.imshow(np.transpose(grid, (1,2,0)).numpy())
    # transpose and convert to rgb 
    images =  np.transpose(grid, (1,2,0))
    # covert BGR to RGB     
    plt.imshow(images)
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
plot_batch(y[:2])
```
